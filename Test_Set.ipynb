```python
# remove '$' and '%" in the string, then convert object to float for column x41 and x45
df_categ['x41'] = df_categ['x41'].str.strip('$')
df_categ['x41'] = df_categ['x41'].astype(float)

df_categ['x45'] = df_categ['x45'].str.strip('%')
df_categ['x45'] = df_categ['x45'].astype(float)
```

```python
# Divide into 4 buckets based on price percentile 
df_categ['x41'] = pd.qcut(df_categ['x41'], q=4, labels=["one", "two", "three", "four"])
```

```python
# Divide into 9 buckets for percentage percentile 

per_labels = {-0.04:'One',-0.03:'Two', -0.02:'Three', -0.01:'Four', 0.00:'Five', 0.01:'Six', 0.02:'Seven', 0.03:'Eight', 0.04:'Nine'}
df_categ['x45'] = df_categ['x45'].map(per_labels)
```

```python
# Impute categorical missing data 
from sklearn.impute import SimpleImputer
impute = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df_categ = pd.DataFrame(impute.fit_transform(df_categ), columns=df_categ.columns)
```

```python
# numerical variables has some outlier, so SimpleImputer (median strategy) might be a better approach
df_num.describe(include=[np.float64])
```

```python
df_categ
```

```python
# Anything not in the range of (Q1 - 1.5 IQR) and (Q3 + 1.5 IQR) is an outlier
# There are 21039 rows out of 40000 rows that has at least one outliers features(52.59%)

Q1 = df_num.quantile(0.25)
Q3 = df_num.quantile(0.75)
IQR = Q3 - Q1
outlier_step = 1.5 * IQR

df_outlier = df_num[~((df_num < (Q1 - outlier_step )) | (df_num > (Q3 + outlier_step))).any(axis=1)]
df_outlier.shape
```

```python
# Impute numerical missing data 
from sklearn.impute import SimpleImputer
impute = SimpleImputer(missing_values=np.nan, strategy='median')
df_num = pd.DataFrame(impute.fit_transform(df_num), columns=df_num.columns)
```

```python
# quick check whether numerical missing data is filled
df_num.isnull().sum().sum()
```

```python
# quick check whether categorical missing data is filled
df_categ.isnull().sum().sum()
```

```python
# create and implement OnehotEncoder
encoder = OneHotEncoder()
onehot_categ = pd.DataFrame(encoder.fit_transform(df_categ).toarray())
onehot_categ
```

```python
# standarize scaling for numerical data 
scaler = StandardScaler()
df_num = pd.DataFrame(scaler.fit_transform(df_num))
df_num.head()
```

```python
# combine categorical data and numerical dataset and assign it to X for later model building
# X = pd.concat([df_num, onehot_categ], axis = 1)
# X

X = pd.concat([df_num, onehot_categ], axis=1, ignore_index=True)
X
```

<div class="alert alert-block alert-warning">
<b><font size="2">Section Summary </font></b> 
    
I separate data into `category` and `numerical` to avoid errors for imputing missing value for both string and float data type, and as well more accurate in imputation by using two strategies `most_frequency` for categorical features and `median` for numerical features when using Univariate feature imputation method due to `52.59%` of the outliers in numerical data (SimpleImputer). 
    
A more sophisticated approach is to use the Multivariate feature imputation. However, it is still in experimental mode, which issues might occur and therefore not recommended as for now.
    
The next step is to transform categories into nominal, so we have all data in numerical form. `One-Hot Encoder` in scik-learn is a better option for nominal compare to other encoder methods like label encoder. Since we only have 4 column features, it won't take that much space as well for the disadvantage of this technique. 
    
After that, I use `standarization feature scaling` for numerical data to have consistant ouput. And lastly, combining categorical data with encoding and numerical data to do further model building and deployment.
