```python
# Impute categorical missing data 
from sklearn.impute import SimpleImputer
impute = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df_categ = pd.DataFrame(impute.fit_transform(df_categ), columns=df_categ.columns)
```

```python
# numerical variables has some outlier, so SimpleImputer (median strategy) might be a better approach
df_num.describe(include=[np.float64])
```

```python
df_categ
```

```python
# Anything not in the range of (Q1 - 1.5 IQR) and (Q3 + 1.5 IQR) is an outlier
# There are 21039 rows out of 40000 rows that has at least one outliers features(52.59%)

Q1 = df_num.quantile(0.25)
Q3 = df_num.quantile(0.75)
IQR = Q3 - Q1
outlier_step = 1.5 * IQR

df_outlier = df_num[~((df_num < (Q1 - outlier_step )) | (df_num > (Q3 + outlier_step))).any(axis=1)]
df_outlier.shape
```

```python
# Impute numerical missing data 
from sklearn.impute import SimpleImputer
impute = SimpleImputer(missing_values=np.nan, strategy='median')
df_num = pd.DataFrame(impute.fit_transform(df_num), columns=df_num.columns)
```

```python
# quick check whether numerical missing data is filled
df_num.isnull().sum().sum()
```

```python
# quick check whether categorical missing data is filled
df_categ.isnull().sum().sum()
```

```python
# assign target y variable to y for further model building
y = df_categ['y']

# convert object type to int
y = y.astype('int')
```

```python
# drop y variable 
df_categ.drop('y', axis=1, inplace= True)
```

```python
# create and implement OnehotEncoder
encoder = OneHotEncoder()
onehot_categ = pd.DataFrame(encoder.fit_transform(df_categ).toarray())
onehot_categ
```

```python
# standarize scaling for numerical data 
scaler = StandardScaler()
df_num = pd.DataFrame(scaler.fit_transform(df_num))
df_num.head()
```

```python
# combine categorical data and numerical dataset and assign it to X for later model building
X = pd.concat([df_num, onehot_categ], axis=1, ignore_index=True)
X
```

```python
# Check both numerical and categorical dataset are consistent
df_num.shape, onehot_categ.shape
```

```python
# Check the data type to see if it's numerical
X.info()
```

<div class="alert alert-block alert-warning">
<b><font size="2">Section Summary </font></b> 
    
I separate data into `category` and `numerical` to avoid errors for imputing missing value for both string and float data type, and as well more accurate in imputation by using two strategies `most_frequency` for categorical features and `median` for numerical features when using Univariate feature imputation method due to `52.59%` of the outliers in numerical data (SimpleImputer). 
    
A more sophisticated approach is to use the Multivariate feature imputation. However, it is still in experimental mode, which issues might occur and therefore not recommended as for now.
    
The next step is to transform categories into nominal, so we have all data in numerical form. `One-Hot Encoder` in scik-learn is a better option for nominal compare to other encoder methods like label encoder. Since we only have 4 column features, it won't take that much space as well for the disadvantage of this technique. 
    
After that, I use `standarization feature scaling` for numerical data to have consistant ouput. And lastly, combining categorical data with encoding and numerical data to do further model building and deployment. 
    
</div>

<div class="alert alert-block alert-success">
<b><font size="4">Model 1: Gradient Boosting with Grid Search </font></b>
<div>

```python
# setup the parameter and use grid search to find the best fit
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier(random_state=0)
param_grid = {'max_features': [2, 5, 10],'n_estimators':[10,20,50],'learning_rate':[0.1,0.25, 0.5,0.75]}

grid = GridSearchCV(gbc, param_grid, cv = 6, n_jobs = -1)
grid.fit(X, y)
```

```python
# Predicted and ouput the class probabilities that the data belonging to the positive class (labeled '1')
probability_gradient_boosting  = grid.predict_proba(df_test)[: ,1]
probability_gradient_boosting 
```

```python
# save the result of the predicted output in csv
np.savetxt("results1.csv", probability_gradient_boosting, delimiter=",")
```

<div class="alert alert-block alert-success">
<b><font size="4">Model 2: Logistic Regression with RFECV</font></b>
<div>

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression
```

```python
# Implementing feature ranking with recursive feature elimination method using logistic regression
# and use cross-validated selection to find the best number of features

rfecv = RFECV(LogisticRegression(), step=1, cv=6, n_jobs = -1, scoring='roc_auc')
rfecv.fit(X, y)
```

```python
# Predicted and ouput the class probabilities that the data belonging to the positive class (labeled '1')
probability_logistic_regression = rfecv.predict_proba(df_test)[: ,1]
probability_logistic_regression 
```

```python
# save the result of the predicted output in csv
np.savetxt("results2.csv", probability_logistic_regression, delimiter=",")
```

<font size="2">

This is a supervised machine learning categorized as a binary classification problem. Some of the common methods used for binary classification including Decision Trees, Support vector machines, Logistic Regression, k-Nearest Neighbors, Naive Bayes, and Neural Networks.
  
It is hard to know right at the start which algorithm will work best, but Logistic Regression is a potentially good starting approach since it is specifically designed for the binary problem. It is also very easy to implement, interpret and very efficient to train. It is the one of the most common algorithm use in the real world. However, Logistic regression attempts to predict outcomes based on a set of independent variables. Therefore, I implemented a feature selection and cross-validation method due to minimize the cons.

I picked the Ensembling Learning method gradient, decision tree boosting as my best choice since it is fast for both training and prediction mostly with good performance. It is also a perfect algorithm for a mix of categorical and continuous data. Even if we changed to the business world, boosted tree algorithm is very commonly used. Although gradient boosting is sensitive to overfitting and noise, we could use the cross-validation method to minimize the cons. Moreover, most libraries today have tools to avoid overfitting.
    
<b>Following are the list of Pros and Cons of the two algorithms:
<b></font>
    
    
Gradient Tree Boosting:
    
`Pros:` 

 - Fast
 - Easy to tune
 - Not sensitive to scale
 - Good performance 
    
`Cons:`
 - Sensitive to overfitting and noise
    
    
Logistic Regression:
    
`Pros:`
    
 - Fast to implement and train
 - Unlikely to overfit
 - Have probability predictions
 - Easy to interpret 
 - Good with large dataset
 
`Cons:`
 - Require Feature Engineering
 - Can’t solve non-linear problems
